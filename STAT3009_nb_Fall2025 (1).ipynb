## Global Mean Implementation Idea

### Global Average

The global mean implementation idea is based on the assumption that the overall rating behavior is consistent across all users and items. The idea is to predict a user's rating for an item based on the global average rating across all users and items.

Mathematically, the global mean can be represented as:

$
		\bar{r} = \frac{1}{|\mathcal{R}|} \sum_{r \in \mathcal{R}} r, \quad \hat{r}_{ui} = \bar{r},
$

where:

* $\bar{r}$ is the global average rating
* $\mathcal{R}$ is the set of all ratings
* $r$ is a rating in the set $\mathcal{R}$
* $\hat{r}_{ui}$ is the predicted rating for user $u$ and item $i$

The implementation steps for the global mean idea are:

### Algorithm

1. **Compute the global average**:
   * Calculate the average rating across all users and items in the training set.
2. **Predict ratings**:
   * Predict the ratings for all user-item pairs in the testing set as the global average.

````python
## GA methods
ga = np.mean(y_train)

n_test = len(X_test)
y_pred = np.ones(n_test)
y_pred = ga*y_pred
````

````python
## package a python function

def ga(X_train, y_train, X_test):
  ## fit the ga param
  ga = np.mean(y_train)
  ## make prediction for the X_test
  n_test = len(X_test)
  y_pred = np.ones(n_test)
  y_pred = ga*y_pred

  return y_pred
````

## User Mean Implementation Idea

### User Average

The user mean implementation idea is based on the assumption that a user's rating behavior is consistent across different items. The idea is to predict a user's rating for an item based on the user's average rating across all items.

Mathematically, the user mean can be represented as:

$
		\bar{r}_{u} = \frac{1}{|\mathcal{I}_u|} \sum_{i \in \mathcal{I}_u} r_{ui}, \text{ for } u=1, \cdots, n; \quad \hat{r}_{ui} = \bar{r}_u
$

where:

* $\bar{r}_{u}$ is the average rating for user $u$
* $\mathcal{I}_u$ is the set of items rated by user $u$
* $r_{ui}$ is the rating given by user $u$ to item $i$
* $\hat{r}_{ui}$ is the predicted rating for user $u$ and item $i$

The implementation steps for the user mean idea are:

### Algorithm

1. **Loop for all users**:
	* Find all records for this user in both training and testing sets.
	* Compute the average ratings for this user in the training set.
	* Predict the ratings for this user in the testing set.

### user average

$$
		\bar{r}_{u} = \frac{1}{|\mathcal{I}_u|} \sum_{i \in \mathcal{I}_u} r_{ui}, \text{ for } u=1, \cdots, n; \quad \hat{r}_{ui} = \bar{r}_u
$$

- Loop for all users
  - Find all records for this user in both training and testing sets.
  - Compute the average ratings for this user in the training set.
  - Predict the ratings for this users in the testing set.

````python
## (InClass Practice) user average

ga = np.mean(y_train)

# Step 1: compute the user mean vector
mu_user = np.zeros(n_user)
for u in range(n_user):
  user_ind = np.where(X_train[:,0] == u)[0]
  if len(user_ind) > 0:
    mu_user[u] = y_train[user_ind].mean()
  else:
    ## user_id has no record in the train dataset
    mu_user[u] = ga
````

````python
## Step 2: make prediction
n_test = len(X_test)
y_pred = np.zeros(n_test)

for k in range(n_test):
  user_tmp = X_test[k,0]
  y_pred[k] = mu_user[user_tmp]
````

````python
## package as a python function

def ua(X_train, y_train, X_test, n_user, min_data=1):
  # Step 1: compute the user mean vector

  # est global mean
  ga = np.mean(y_train)
  # est user mean
  mu_user = np.zeros(n_user)

  for u in range(n_user):
    user_ind = np.where(X_train[:,0] == u)[0]
    if len(user_ind) > min_data:
      mu_user[u] = y_train[user_ind].mean()
    else:
      ## user_id has no record in the train dataset
      mu_user[u] = ga

  # make prediction
  n_test = len(X_test)
  y_pred = np.zeros(n_test)

  for k in range(n_test):
    user_tmp = X_test[k,0]
    y_pred[k] = mu_user[user_tmp]

  return y_pred
````

## Item Mean Implementation Idea

### Item Average

The item mean implementation idea is based on the assumption that an item's rating behavior is consistent across different users. The idea is to predict a user's rating for an item based on the item's average rating across all users.

Mathematically, the item mean can be represented as:

$
		\bar{r}_{i} = \frac{1}{|\mathcal{U}_i|} \sum_{u \in \mathcal{U}_i} r_{ui}, \text{ for } i=1, \cdots, m; \quad \hat{r}_{ui} = \bar{r}_i,
$

where:

* $\bar{r}_{i}$ is the average rating for item $i$
* $\mathcal{U}_i$ is the set of users who rated item $i$
* $r_{ui}$ is the rating given by user $u$ to item $i$
* $\hat{r}_{ui}$ is the predicted rating for user $u$ and item $i$

The implementation steps for the item mean idea are:

### Algorithm

1. **Loop for all items**:
	* Find all records for this item in both training and testing sets.
	* Compute the average ratings for this item in the training set.
	* Predict the ratings for this item in the testing set.

````python
## (InClass Practice) item average
## package as a python function

def ia(X_train, y_train, X_test, n_item, min_data=1):
  # Step 1: compute the item mean vector

  # est global mean
  ga = np.mean(y_train)
  # est item mean
  mu_item = np.zeros(n_item)

  for i in range(n_item):
    item_ind = np.where(X_train[:,1] == i)[0]
    if len(item_ind) > min_data:
      mu_item[i] = y_train[item_ind].mean()
    else:
      ## user_id has no record in the train dataset
      mu_item[i] = ga

  # make prediction
  n_test = len(X_test)
  y_pred = np.zeros(n_test)

  for k in range(n_test):
    item_tmp = X_test[k,1]
    y_pred[k] = mu_item[item_tmp]

  return y_pred
````

````python
## report RMSE

y_pred = ia(X_train,
            y_train,
            X_test,
            n_item,
            min_data=0)

rmse(y_test, y_pred)
````

## Package Python functions

In this section, we will package the Python functions for the baseline models implemented earlier. This will make it easier to reuse and test these functions.


````python
## Package Python functions for simple recommender systems

# Global Mean Function
def glb_mean(train_rating, test_pair):
    """
    Predict ratings based on the global mean of the training ratings.

    Parameters:
    train_rating (array): Array of training ratings
    test_pair (array): Array of user-item pairs for testing

    Returns:
    pred (array): Predicted ratings
    """
    pred = train_rating.mean() * np.ones(len(test_pair))
    return pred

# User Mean Function
def user_mean(train_pair, train_rating, test_pair):
    """
    Predict ratings based on the mean rating for each user.

    Parameters:
    train_pair (array): Array of user-item pairs for training
    train_rating (array): Array of training ratings
    test_pair (array): Array of user-item pairs for testing

    Returns:
    pred (array): Predicted ratings
    """
    n_user = max(train_pair[:,0].max(), test_pair[:,0].max())+1
    pred = np.zeros(len(test_pair))
    glb_mean_value = train_rating.mean()

    # Loop through each user
    for u in range(n_user):
        # Find the indices for both train and test for user_id = u
        ind_test = np.where(test_pair[:,0] == u)[0]
        ind_train = np.where(train_pair[:,0] == u)[0]

        if len(ind_test) == 0:
            continue
        if len(ind_train) < 3:
            # If user has less than 3 ratings, use global mean
            pred[ind_test] = glb_mean_value
        else:
            # Predict as user average
            pred[ind_test] = train_rating[ind_train].mean()

    return pred

# Item Mean Function
def item_mean(train_pair, train_rating, test_pair):
    """
    Predict ratings based on the mean rating for each item.

    Parameters:
    train_pair (array): Array of user-item pairs for training
    train_rating (array): Array of training ratings
    test_pair (array): Array of user-item pairs for testing

    Returns:
    pred (array): Predicted ratings
    """
    n_item = max(train_pair[:,1].max(), test_pair[:,1].max())+1
    pred = np.zeros(len(test_pair))
    glb_mean_value = train_rating.mean()

    # Loop through each item
    for i in range(n_item):
        # Find the indices for both train and test for item_id = i
        ind_test = np.where(test_pair[:,1] == i)[0]
        ind_train = np.where(train_pair[:,1] == i)[0]

        if len(ind_test) == 0:
            continue
        if len(ind_train) < 3:
            # If item has less than 3 ratings, use global mean
            pred[ind_test] = glb_mean_value
        else:
            # Predict as item average
            pred[ind_test] = train_rating[ind_train].mean()

    return pred
````

## Hybrid models; user-item mean


In this section, we will explore a hybrid model that combines the strengths of both user-based and item-based models. Specifically, we will implement a model that predicts ratings by combining the user mean and item mean.

The predicted rating $\hat{r}_{ui}$ is calculated as the sum of the user mean $\hat{\mu}_u$ and item mean $\hat{\mu}_i$:

$$\hat{r}_{ui} = \mu_u + \mu_i$$

where
		$$\mu_u = \frac{1}{|\mathcal{I}_u|} \sum_{i \in \mathcal{I}_u} r_{ui}, \quad \mu_i = \frac{1}{|\mathcal{U}_i|} \sum_{u \in \mathcal{U}_i} (r_{ui} - \mu_u)$$

````python
## GA + UA + IA + ...

## y ~ ga
y_pred = ga(X_train, y_train, X_test)

## y - ga ~ ua
delta1 = y_train - ga(X_train, y_train, X_train)
y_pred = y_pred + ua(X_train, delta1, X_test, n_user)

## y - ga - ua ~ ia
delta2 = delta1 - ua(X_train, delta1, X_train, n_user)
y_pred = y_pred + ia(X_train, delta2, X_test, n_item)

## NEW DATA
rmse(y_test, y_pred)
````

````python
from sklearn.base import BaseEstimator

## global average RS

class glbAve(BaseEstimator):
  def __init__(self, _glb_mean=0.0):
    self._glb_mean = _glb_mean

  def fit(self, X, y):
    ## update (fitting) params
    self.is_fitted_ = True
    self._glb_mean = np.mean(y)
    return self

  def predict(self, X):
    ## how to make prediction
    y = np.ones(len(X))
    return y * self._glb_mean
````

````python
class userAve(BaseEstimator):
  def __init__(self, n_user, min_data=0):
    ## setup params
    self.n_user = n_user
    ## fitting params
    self._glb_mean = 0.0
    self._user_mu = np.zeros(n_user)
    ## hyper params
    self.min_data = min_data

  def fit(self, X, y):
    ## fit glb mean param
    self._glb_mean = np.mean(y)

    ## fit user mean param
    for user_tmp in range(self.n_user):
      user_index = np.where(X[:,0] == user_tmp)[0]
      ## (nearly) cold-start users
      if len(user_index) <= self.min_data:
        self._user_mu[user_tmp] = self._glb_mean
      else:
        self._user_mu[user_tmp] = np.mean(y[user_index])
    return self

  def predict(self, X):
    user_index = X[:,0]
    return self._user_mu[user_index]
````

````python
class itemAve(BaseEstimator):
  def __init__(self, n_item, min_data=0):
    ## setup params
    self.n_item = n_item
    ## fitting params
    self._glb_mean = 0.0
    self._item_mu = np.zeros(n_item)
    ## hyper params
    self.min_data = min_data

  def fit(self, X, y):
    ## fit glb mean param
    self._glb_mean = np.mean(y)

    ## fit item mean param
    for item_tmp in range(self.n_item):
      item_index = np.where(X[:,1] == item_tmp)[0]
      ## (nearly) cold-start users
      if len(item_index) <= self.min_data:
        self._item_mu[item_tmp] = self._glb_mean
      else:
        self._item_mu[item_tmp] = np.mean(y[item_index])
    return self

  def predict(self, X):
    item_index = X[:,1]
    return self._item_mu[item_index]
````

````python
## Below is the code loading the dataset

import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

train = pd.read_csv('https://raw.githubusercontent.com/statmlben/CUHK-STAT3009/main/dataset/netflix/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/statmlben/CUHK-STAT3009/main/dataset/netflix/test.csv')

## RS data casting with ML format
X_train = train[['user_id', 'movie_id']].values
y_train = train['rating'].values

X_test = test[['user_id', 'movie_id']].values
y_test = test['rating'].values

n_users = len(set(X_train[:,0]).union(X_test[:,0]))
n_items = len(set(X_train[:,1]).union(X_test[:,1]))
````

**`__init__`**: This is the constructor method that initializes the object. It takes in several parameters:

* `n_user`: The number of users in the system.
* `n_item`: The number of items in the system.
* `lam`: The regularization parameter (default is 0.001).
* `K`: The number of latent factors (default is 10).
* `iterNum`: The number of iterations for the ALS algorithm (default is 10).
* `tol`: The tolerance for convergence (default is 1e-4).
* `verbose`: A flag to control the verbosity of the algorithm (default is 1).

The constructor initializes the `P` and `Q` matrices, which represent the user and item latent factors, respectively.

````python
import numpy as np
from sklearn.base import BaseEstimator
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

class SVD(BaseEstimator):
  def __init__(self, n_user, n_item,
                     K=5, lam=0.01,
                     iterNum=10, tol=1e-4):
    ## data params
    self.n_user = n_user
    self.n_item = n_item
    ## hp
    self.K = K ## num of latent factors
    self.lam = lam ## reg weight of l2 penalty
    ## params
    self.P = np.random.randn(n_user, K)
    self.Q = np.random.randn(n_item, K)
    ## fitting params
    self.iterNum = iterNum
    self.tol = tol

  def fit(self, X, y):
    n_obs = len(X)
    diff = 1.0

    obj_tmp = self.obj(X, y)

    for h in range(self.iterNum):
      ## termination condition
      if diff < self.tol:
        break
      ## P fixed; solve Q
      for i in range(self.n_item):
        ## solve qi
        index_tmp = np.where(X[:,1] == i)[0]
        ## cold-start items
        if len(index_tmp) == 0:
          self.Q[i] = np.zeros(self.K)
        else:
          y_tmp = y[index_tmp]
          P_tmp = self.P[X[index_tmp,0]]
          clf_tmp = Ridge(alpha=self.lam*n_obs,
                          fit_intercept=False)
          clf_tmp.fit(X=P_tmp, y=y_tmp)
          self.Q[i] = clf_tmp.coef_

      ## Q is fixed; solve P
      for u in range(self.n_user):
        ## solve pu
        index_tmp = np.where(X[:,0] == u)[0]
        ## cold-start users
        if len(index_tmp) == 0:
          self.P[u] = np.zeros(self.K)
        else:
          y_tmp = y[index_tmp]
          Q_tmp = self.Q[X[index_tmp,1]]
          clf_tmp = Ridge(alpha=self.lam*n_obs,
                          fit_intercept=False)
          clf_tmp.fit(X=Q_tmp, y=y_tmp)
          self.P[u] = clf_tmp.coef_

      ## compute obj, diff
      obj_old = obj_tmp
      obj_tmp = self.obj(X, y)
      diff = obj_old - obj_tmp

      print(f"SVD-BCD: step: {h}; obj: {obj_tmp}, diff: {diff}")
    return self

  def predict(self, X):
    ## X: [user_id, item_id]
    y_pred = []
    for x_tmp in X:
      user_id_tmp = x_tmp[0]
      item_id_tmp = x_tmp[1]

      y_pred_tmp = self.P[user_id_tmp] @ self.Q[item_id_tmp]
      y_pred.append(y_pred_tmp)
    return np.array(y_pred)

  def obj(self, X, y):
    y_pred = self.predict(X)
    mse = mean_squared_error(y,y_pred)
    pen = np.sum(self.P**2) + np.sum(self.Q**2)
    return mse + self.lam * pen
````


**`predict`**: This method predicts the ratings for a given set of user-item pairs. It takes in a 2D array `X` of shape `(n_samples, 2)`, where each row represents a user-item pair. It returns a 1D array of shape `(n_samples,)`, where each element represents the predicted rating.

**`mse`**: This method computes the mean squared error (MSE) between the predicted ratings and the actual ratings. It takes in two parameters:

* `X`: A 2D array of shape `(n_samples, 2)`, where each row represents a user-item pair.
* `y`: A 1D array of shape `(n_samples,)`, where each element represents the actual rating.

It returns the MSE value.

**`obj`**: This method computes the objective function value, which is the sum of the MSE and the regularization term. It takes in two parameters:

* `X`: A 2D array of shape `(n_samples, 2)`, where each row represents a user-item pair.
* `y`: A 1D array of shape `(n_samples,)`, where each element represents the actual rating.

It returns the objective function value.

**`fit`**: This method fits the matrix factorization model to the given data. It takes in two parameters:

* `X`: A 2D array of shape `(n_samples, 2)`, where each row represents a user-item pair.
* `y`: A 1D array of shape `(n_samples,)`, where each element represents the rating associated with the corresponding user-item pair.

The method performs the following steps:

1. Initializes the `index_item` and `index_user` arrays, which store the indices of the user-item pairs for each item and user, respectively.
2. Performs the ALS algorithm for the specified number of iterations.
3. In each iteration, it updates the item and user latent factors using the current estimates of the other factor.
4. Computes the objective function value and checks for convergence.
5. Returns the fitted model.

Note that the `fit` method is the main method that performs the matrix factorization, and the other methods are used to compute the predicted ratings, MSE, and objective function value.

How to implement the following model:
$$
f(\mathbf{x}) = \psi_{[3, 1]}(\mathbf{x}) + \psi_{[2,1]}(\mathbf{x})
$$

- The point is to define `layer` and **PATH** separately

````python
from tensorflow import keras
from keras.layers import Dense

class pNet(keras.Model):

  def __init__(self):
    super().__init__()
    ## define all layers here
    self.dense1 = Dense(units=3, activation='relu')
    self.dense2 = Dense(units=2, activation='relu')
    self.dense3 = Dense(units=1)
    self.dense4 = Dense(units=1)

  def call(self, inputs):
    ## inputs -> outputs
    # up panel
    up = self.dense1(inputs)
    up = self.dense3(up)
    # low panel
    low = self.dense2(inputs)
    low = self.dense4(low)
    # out
    out = up + low
    return out
````

- Implement the following model using Keras:

$$
f(\mathbf{x}) = \psi_{[3, 1]}(\mathbf{A} \mathbf{x}) + \psi_{[2,1]}( \mathbf{A} \mathbf{x}),
$$

where $\mathbf{A} \in \mathbb{R}^{5 \times 8}$ is a parameter matrix.

- **Hint**: The operation $\mathbf{A} \mathbf{x}$ can be implemented as a Dense layer with 5 nodes and a `linear` activation function.

````python
from tensorflow import keras
from keras.layers import Dense

class linpNet(keras.Model):

  def __init__(self):
    super().__init__()
    ## define all layers here
    self.denseA = Dense(units=5, use_bias=False)
    self.dense1 = Dense(units=3, activation='relu')
    self.dense2 = Dense(units=2, activation='relu')
    self.dense3 = Dense(units=1)
    self.dense4 = Dense(units=1)

  def call(self, inputs):
    ## inputs -> outputs
    # linA panel
    inputs = self.denseA(inputs) # Ax
    # up panel
    up = self.dense1(inputs)
    up = self.dense3(up)
    # low panel
    low = self.dense2(inputs)
    low = self.dense4(low)
    # out
    out = up + low
    return out
````

````python
## compile model
## opt: SGD
## Loss: MAE
## Metrics: RMSE and MAPE

test_model = linpNet()

# opt
opt = keras.optimizers.SGD(1e-4)

test_model.compile(optimizer=opt,
                  loss='mean_squared_error',
                  metrics=['root_mean_squared_error',
                       'mean_absolute_error'])

test_model.fit(x=X_train,
               y=y_train,
               batch_size=64,
               epochs=20)
````

### Define LFM by `tf.keras`
- Define the layers: embedding layers: embed both users and movies in to 50-dimensional vectors.
- Connect from `input` to `output`: LFM computes a match score between user and movie embeddings via a dot product, and adds a per-movie and per-user bias.

Take a close look to [tf.keras.layers.Embedding](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Embedding)

    tf.keras.layers.Embedding(
        input_dim,
        output_dim,
        embeddings_initializer='uniform',
        embeddings_regularizer=None,
        activity_regularizer=None,
        embeddings_constraint=None,
        mask_zero=False,
        input_length=None,

    )

- `input_dim`: Integer. Size of the vocabulary, i.e. maximum integer index + 1.

- `output_dim`: Integer. Dimension of the dense embedding.

- `embeddings_initializer`: Initializer for the embeddings matrix (see keras.initializers).

- `embeddings_regularizer`: Regularizer function applied to the embeddings matrix (see keras.regularizers).

- `embeddings_constraint`: Constraint function applied to the embeddings matrix (see keras.constraints).

- `mask_zero`: Boolean, whether or not the input value 0 is a special "padding" value that should be masked out. This is useful when using recurrent layers which may take variable length input. If this is True, then all subsequent layers in the model need to support masking or an exception will be raised. If mask_zero is set to True, as a consequence, index 0 cannot be used in the vocabulary (input_dim should equal size of vocabulary + 1).

- `input_length`: Length of input sequences, when it is constant. This argument is required if you are going to connect Flatten then Dense layers upstream (without it, the shape of the dense outputs cannot be computed).

````python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from pathlib import Path
import matplotlib.pyplot as plt
````

````python
class NMF(keras.Model):
  def __init__(self, num_user, num_item, K=10):
    super().__init__()
    ## layers
    self.user_embed = layers.Embedding(input_dim=num_user,
                                       output_dim=K)
    self.item_embed = layers.Embedding(input_dim=num_item,
                                       output_dim=K)
    self.dot = layers.Dot(axes=1)
    # Specify the axis to sum over

  def call(self, inputs):
    # pu: B x r
    user_vec = self.user_embed(inputs[:,0])
    # qi: B x r
    item_vec = self.item_embed(inputs[:,1])

    out = self.dot([user_vec, item_vec])

    # out = tf.reduce_sum(user_vec*item_vec,
    #                     axis=1,
    #                     keepdims=True)
    return out
````

````python
class BNMF(keras.Model):
  def __init__(self, num_user, num_item, K=10):
    super().__init__()
    ## layers
    self.user_embed = layers.Embedding(input_dim=num_user,
                                       output_dim=K)
    self.item_embed = layers.Embedding(input_dim=num_item,
                                       output_dim=K)
    self.user_bias = layers.Embedding(input_dim=num_user,
                                       output_dim=1)
    self.item_bias = layers.Embedding(input_dim=num_item,
                                       output_dim=1)
    self.dot = layers.Dot(axes=1)
    # Specify the axis to sum over

  def call(self, inputs):
    # pu: B x r
    user_vec = self.user_embed(inputs[:,0])
    # qi: B x r
    item_vec = self.item_embed(inputs[:,1])
    # au: B x 1
    user_const = self.user_bias(inputs[:,0])
    # bi: B x 1
    item_const = self.item_bias(inputs[:,1])

    ## pu^T q_i + au + bi
    out = self.dot([user_vec, item_vec]) + user_const + item_const

    # out = tf.reduce_sum(user_vec*item_vec,
    #                     axis=1,
    #                     keepdims=True)
    return out
````

````python
class NCF(keras.Model):
  def __init__(self, num_user, num_item, K=10):
    super().__init__()
    ## layers
    self.user_embed = layers.Embedding(input_dim=num_user,
                                       output_dim=K)
    self.item_embed = layers.Embedding(input_dim=num_item,
                                       output_dim=K)
    self.concat = layers.Concatenate()

    self.dense1 = layers.Dense(K, activation='relu')
    self.dense2 = layers.Dense(4, activation='relu')
    self.out = layers.Dense(1)

  def call(self, inputs):
    # pu: B x r
    user_vec = self.user_embed(inputs[:,0])
    # qi: B x r
    item_vec = self.item_embed(inputs[:,1])

    joint_vec = self.concat([user_vec, item_vec])

    dense_vec = self.dense1(joint_vec)
    dense_vec = self.dense2(dense_vec)
    out = self.out(dense_vec)

    return out
````

## Additive NCF (A-NCF):

- Recall the figure
- Define the layers: layers.Embedding + layers.concatenate + layers.Dense
- Connect from input to output...

````python
class ANCF(keras.Model):
  def __init__(self, num_user, num_item, K=10, R=5):
    super().__init__()
    ## layers
    # MF embdding
    self.user_embed = layers.Embedding(input_dim=num_user,
                                       output_dim=K)
    self.item_embed = layers.Embedding(input_dim=num_item,
                                       output_dim=K)
    # Dense FC embdding
    self.user_fc_embed = layers.Embedding(input_dim=num_user,
                                       output_dim=R)
    self.item_fc_embed = layers.Embedding(input_dim=num_item,
                                       output_dim=R)


    self.concat = layers.Concatenate()
    self.dot = layers.Dot(axes=1)

    self.dense1 = layers.Dense(R, activation='relu')
    self.dense2 = layers.Dense(4, activation='relu')
    self.out = layers.Dense(1)

  def call(self, inputs):
    ## r_{ui} = p^T_u q_i + \phi( [s_u, t_i] )

    ## MF
    # pu: B x K
    user_vec = self.user_embed(inputs[:,0])
    # qi: B x K
    item_vec = self.item_embed(inputs[:,1])
    out_mf = self.dot([user_vec, item_vec])

    ## FC
    # su: B x R
    user_fc_vec = self.user_fc_embed(inputs[:,0])
    # ti: B x R
    item_fc_vec = self.item_fc_embed(inputs[:,1])
    joint_vec = self.concat([user_fc_vec, item_fc_vec])

    dense_vec = self.dense1(joint_vec)
    dense_vec = self.dense2(dense_vec)
    out_fc = self.out(dense_vec)

    return out_mf + out_fc
````

````python
model = ANCF(n_users, n_items)
````

````python
opt = keras.optimizers.Adam(1e-3)

model.compile(optimizer=opt,
              loss='mean_squared_error',
              metrics=['root_mean_squared_error'])

## callback
callback = [keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    mode='min',
    restore_best_weights=True,
)]

## fit with early stopping
model.fit(x=X_train,
             y=y_train,
             epochs=20,
             batch_size=512,
             validation_split=0.3,
             callbacks=callback
             )
````

## **LinearRS Neural Network Model**

**Overview**

The LinearRS model is a neural network architecture designed for recommender systems. It takes in user and item features as input and outputs a predicted rating or engagement score.

**Network Structure**

The LinearRS model consists of the following components:

### Embedding Layers

* **User Embedding Layer**: maps user categorical features (e.g. personId, userAgent, userRegion) to a dense vector representation
* **Item Embedding Layer**: maps item categorical features (e.g. contentId, authorPersonId, contentType, lang) to a dense vector representation
* **Agent Embedding Layer**: maps userAgent features to a dense vector representation
* **Region Embedding Layer**: maps userRegion features to a dense vector representation
* **Author Embedding Layer**: maps authorPersonId features to a dense vector representation
* **ContentType Embedding Layer**: maps contentType features to a dense vector representation
* **Lang Embedding Layer**: maps lang features to a dense vector representation

### Dense Layers

* **User Dense Layer**: maps user dense features (e.g. count_user, mean_user) to a dense vector representation
* **Item Dense Layer**: maps item dense features (e.g. count_author, mean_author) to a dense vector representation

### Output Layer

* The output layer computes the dot product of the user and item vector representations and returns a predicted rating or engagement score.

**Code Explanation**
-------------------

The code defines the LinearRS model using the Keras API. The model takes in four input tensors:

* `UD_feat`: user dense features
* `ID_feat`: item dense features
* `UC_feat`: user categorical features
* `IC_feat`: item categorical features

The model consists of several embedding layers and dense layers, which are defined using the `layers.Embedding` and `layers.Dense` classes from Keras.

The `call` method defines the forward pass through the network. It first embeds the user and item categorical features using the respective embedding layers. Then, it maps the user and item dense features using the respective dense layers. Finally, it computes the dot product of the user and item vector representations and returns the predicted rating or engagement score.

**Hyperparameters**
-------------------

* `num_users`: number of unique users
* `num_items`: number of unique items
* `num_agent`: number of unique user agents
* `num_region`: number of unique user regions
* `num_author`: number of unique authors
* `num_ctype`: number of unique content types
* `num_lang`: number of unique languages
* `embedding_size`: dimensionality of the embedding space

These hyperparameters are used to define the shape of the embedding layers and the dense layers.

````python
# user_cate = ['personId', 'userAgent', 'userRegion']
# item_cate = ['contentId', 'authorPersonId', 'contentType', 'lang']

# user_dense = ['count_user', 'mean_user']
# item_dense = ['count_author', 'mean_author']

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

class LinearRS(keras.Model):
    def __init__(self, num_users, num_items, num_agent, num_region,
                 num_author, num_ctype, num_lang,
                 embedding_size, **kwargs):
        super(LinearRS, self).__init__(**kwargs)

        self.num_users = num_users
        self.num_items = num_items
        self.num_region = num_region
        self.num_author = num_author
        self.num_ctype = num_ctype
        self.num_lang = num_lang
        self.embedding_size = embedding_size

        ## Embedding layers
        self.user_embedding = layers.Embedding(
            num_users,
            embedding_size,
        )
        self.item_embedding = layers.Embedding(
            num_items,
            embedding_size,
        )
        self.agent_embedding = layers.Embedding(
            num_agent,
            embedding_size,
        )
        self.region_embedding = layers.Embedding(
            num_region,
            embedding_size,
        )
        self.author_embedding = layers.Embedding(
            num_author,
            embedding_size,
        )
        self.ctype_embedding = layers.Embedding(
            num_ctype,
            embedding_size,
        )
        self.lang_embedding = layers.Embedding(
            num_lang,
            embedding_size,
        )

        ## Dense layers
        self.user_dense = layers.Dense(embedding_size,
                                       activation='relu')
        self.item_dense = layers.Dense(embedding_size,
                                       activation='relu')

        ## Dot layers
        self.dot = layers.Dot(axes=1)

    def call(self, inputs):
        ## multiple source inputs: cate/dense user/item
        ## inputs = [ds1, ds2, ds3, ... dsn]
        X_UD, X_ID, X_UC, X_IC = inputs[0], inputs[1], inputs[2], inputs[3]

        ## user embedding
        ## ['personId', 'userAgent', 'userRegion']
        user_vec = self.user_embedding(X_UC[:,0])
        agent_vec = self.agent_embedding(X_UC[:,1])
        region_vec = self.region_embedding(X_UC[:,2])

        ## item embedding
        ## ['contentId', 'authorPersonId', 'contentType', 'lang']
        item_vec = self.item_embedding(X_IC[:,0])
        author_vec = self.author_embedding(X_IC[:,1])
        ctype_vec = self.ctype_embedding(X_IC[:,2])
        lang_vec = self.lang_embedding(X_IC[:,3])

        ## user dense
        user_dense_vec = self.user_dense(X_UD)
        item_dense_vec = self.item_dense(X_ID)

        ## p(u, x_u)
        p_vec = user_vec + agent_vec + region_vec + user_dense_vec
        ## q(i, z_i)
        q_vec = item_vec + author_vec + ctype_vec + lang_vec + item_dense_vec

        out = self.dot([p_vec, q_vec])

        return out
````

## Model Build:

**TTowerRS**: Two-Tower models in RS wth Side Information

````python
# user_cate = ['personId', 'userAgent', 'userRegion']
# item_cate = ['contentId', 'authorPersonId', 'contentType', 'lang']

# user_dense = ['count_user', 'mean_user']
# item_dense = ['count_author', 'mean_author']

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

class DTowerRS(keras.Model):
    def __init__(self, num_users, num_items, num_agent, num_region,
                 num_author, num_ctype, num_lang,
                 embedding_size, dense_size, **kwargs):
        super(DTowerRS, self).__init__(**kwargs)
        self.num_users = num_users
        self.num_items = num_items
        self.num_region = num_region
        self.num_author = num_author
        self.num_ctype = num_ctype
        self.num_lang = num_lang
        self.embedding_size = embedding_size
        self.dense_size = dense_size

        ## Embedding layers
        self.user_embedding = layers.Embedding(
            num_users,
            embedding_size,
        )
        self.item_embedding = layers.Embedding(
            num_items,
            embedding_size,
        )
        self.agent_embedding = layers.Embedding(
            num_agent,
            embedding_size,
        )
        self.region_embedding = layers.Embedding(
            num_region,
            embedding_size,
        )
        self.author_embedding = layers.Embedding(
            num_author,
            embedding_size,
        )
        self.ctype_embedding = layers.Embedding(
            num_ctype,
            embedding_size,
        )
        self.lang_embedding = layers.Embedding(
            num_lang,
            embedding_size,
        )

        ## More Mapping layers
        self.user_dense = layers.Dense(dense_size,
                                       name='user_dense',
                                       activation='relu')
        self.item_dense = layers.Dense(dense_size,
                                       name='item_dense',
                                       activation='relu')
        self.joint_dense = layers.Dense(dense_size,
                                        name='joint_dense',
                                        activation='relu')
        self.out_dense = layers.Dense(1, name='out_dense',
                                      activation='linear')

    def call(self, inputs):
        ## multi source inputs: cate/cont data
        UD_feat, ID_feat, UC_feat, IC_feat = inputs[0], inputs[1], inputs[2], inputs[3]

        ## user embedding
        ## ['personId', 'userAgent', 'userRegion']
        user_vec = self.user_embedding(UC_feat[:,0])
        agent_vec = self.agent_embedding(UC_feat[:,1])
        region_vec = self.region_embedding(UC_feat[:,2])

        ## item embedding
        ## ['contentId', 'authorPersonId', 'contentType', 'lang']
        item_vec = self.item_embedding(IC_feat[:,0])
        author_vec = self.author_embedding(IC_feat[:,1])
        ctype_vec = self.ctype_embedding(IC_feat[:,2])
        lang_vec = self.lang_embedding(IC_feat[:,3])

        ## concate all feats
        user_all_vec = layers.Concatenate()([UD_feat,
                                             user_vec,
                                             agent_vec,
                                             region_vec])
        item_all_vec = layers.Concatenate()([ID_feat,
                                             item_vec,
                                             author_vec,
                                             ctype_vec,
                                             lang_vec])

        user_dense_vec = self.user_dense(user_all_vec)
        item_dense_vec = self.item_dense(item_all_vec)

        ## joint dense
        joint_vec = layers.Concatenate()([user_dense_vec, item_dense_vec])
        out = self.joint_dense(joint_vec)
        out = self.out_dense(joint_vec)
        return out
````

````python
## define model
model = DTowerRS(num_users=num_users,
                 num_items=num_items,
                 num_agent=num_agent,
                 num_region=num_region,
                 num_author=num_author,
                 num_ctype=num_ctype,
                 num_lang=num_lang,
                 embedding_size=100,
                 dense_size=128)

metrics = [
    keras.metrics.MeanAbsoluteError(name='mae'),
    keras.metrics.RootMeanSquaredError(name='rmse')
]

model.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss=tf.keras.losses.MeanSquaredError(),
    metrics=metrics
)
````

````python
## model training
callbacks = [keras.callbacks.EarlyStopping(
    monitor='val_rmse', min_delta=0, patience=5, verbose=1,
    mode='min', baseline=None, restore_best_weights=True)]

history = model.fit(
    x=[UD_train, ID_train, UC_train, IC_train],
    y=y_train,
    batch_size=64,
    epochs=50,
    verbose=1,
    validation_split=.2,
    callbacks=callbacks
)
````
